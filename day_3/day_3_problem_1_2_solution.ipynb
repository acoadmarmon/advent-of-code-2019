```python
import numpy as np
import math

#Part 1
n = 25000
def manhattan_distance(coord1, coord2):
    x1, y1 = coord1
    x2, y2 = coord2
    return abs(x1 - x2) + abs(y1 - y2)

def create_line(start_pos, direction, total_distance, distance, arr, mask, additive=1):
    x, y = start_pos
    new_pos = start_pos
    if direction == 'R':
        new_pos = (x + distance, y)
        mask[x :x + distance, y] = np.minimum(mask[x:x + distance, y], np.arange(total_distance, total_distance + distance))
        x = x + 1
        arr[x :x + distance, y] = arr[x:x + distance, y] + additive
    elif direction == 'L':
        new_pos = (x - distance, y)
        mask[x - distance:x, y] = np.minimum(mask[x - distance:x, y], np.arange(total_distance + distance, total_distance, -1))
        x = x - 1
        arr[x - distance:x, y] = arr[x - distance:x, y] + additive
    elif direction == 'U':
        new_pos = (x, y - distance)
        mask[x, y - distance:y] = np.minimum(mask[x, y - distance:y], np.arange(total_distance + distance, total_distance, -1))
        y = y - 1
        arr[x, y - distance:y] = arr[x, y - distance:y] + additive
    elif direction == 'D':
        new_pos = (x, y + distance)
        mask[x, y:y + distance] = np.minimum(mask[x, y:y + distance], np.arange(total_distance, total_distance + distance))
        y = y + 1
        arr[x, y:y + distance] = arr[x, y:y + distance] + additive
    return (arr, new_pos, mask)

def build_circuit(line_list, line_space, mask):
    curr_pos = (int(n/2), int(n/2))
    total_distance = 0
    for i in line_list:
        direction, distance = (i[0], int(i[1:]))
        line_space, new_pos, mask = create_line(curr_pos, direction, total_distance, distance, line_space, mask)
        total_distance = total_distance + distance
        curr_pos = new_pos
    line_space[line_space > 1] = 1
    return (line_space, mask)

def find_nearest_intersection(line_space):
    coord_tuple = np.where(line_space == 2)
    coord_tuple = zip(coord_tuple[0], coord_tuple[1])
    lowest_dist = n
    lowest_coords = (int(n/2), int(n/2))
    for i in coord_tuple:
        dist_from_origin = manhattan_distance(i, (int(n/2), int(n/2))) 
        if dist_from_origin < lowest_dist:
            lowest_dist = dist_from_origin
            lowest_coords = i
    return lowest_dist

find_nearest_intersection(np.add(build_circuit(line1, np.zeros((n, n)), np.full((n,n), np.inf))[0], build_circuit(line2, np.zeros((n, n)), np.full((n,n), np.inf))[0]))
```
