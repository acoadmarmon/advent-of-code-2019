```python
def get_val(arr, position, parameter=0):
    if parameter:
        return arr[position]
    else:
        return arr[arr[position]]

def intcode_parser(arr, pos, input_val, output_val, phase_setting):
    #print(arr, pos, input_val, output_val, phase_setting)
    op = arr[pos]
    parameters = (0, 0, 0)
    if op > 99:
        parameters = ((op // 100) % 10, (op // 1000) % 10, (op // 10000) % 10)
        op = op % 100
        
    if op == 99:
        return output_val
    elif op == 1:
        arr[arr[pos + 3]] = get_val(arr, pos + 1, parameters[0]) + get_val(arr, pos + 2, parameters[1])
        return intcode_parser(arr, pos + 4, input_val, output_val, phase_setting)
    elif op == 2:
        arr[arr[pos + 3]] = get_val(arr, pos + 1, parameters[0]) * get_val(arr, pos + 2, parameters[1])
        return intcode_parser(arr, pos + 4, input_val, output_val, phase_setting)
    elif op == 3:
        arr[arr[pos + 1]] = phase_setting
        return intcode_parser(arr, pos + 2, input_val, output_val, input_val)
    elif op == 4:
        return intcode_parser(arr, pos + 2, input_val, get_val(arr, pos + 1, parameters[0]), phase_setting)
    elif op == 5:
        if get_val(arr, pos + 1, parameters[0]) != 0:
            return intcode_parser(arr, get_val(arr, pos + 2, parameters[1]), input_val, output_val, phase_setting)
        else:
            return intcode_parser(arr, pos + 3, input_val, output_val, phase_setting)
    elif op == 6:
        if get_val(arr, pos + 1, parameters[0]) == 0:
            return intcode_parser(arr, get_val(arr, pos + 2, parameters[1]), input_val, output_val, phase_setting)
        else:
            return intcode_parser(arr, pos + 3, input_val, output_val, phase_setting)
    elif op == 7:
        if get_val(arr, pos + 1, parameters[0]) < get_val(arr, pos + 2, parameters[1]):
            arr[arr[pos + 3]] = 1
        else:
            arr[arr[pos + 3]] = 0
        return intcode_parser(arr, pos + 4, input_val, output_val, phase_setting)
    elif op == 8:
        if get_val(arr, pos + 1, parameters[0]) == get_val(arr, pos + 2, parameters[1]):
            arr[arr[pos + 3]] = 1
        else:
            arr[arr[pos + 3]] = 0
        return intcode_parser(arr, pos + 4, input_val, output_val, phase_setting)
    else:
        print("Uknown op code " + str(op) + "at position " + str(pos))
        return False
```
